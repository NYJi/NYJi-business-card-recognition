```python
import cv2
import numpy as np


img1 = "C:/Users/Yunji Noh/OneDrive/image_processing/data/BC1.jpg"
img2= "C:/Users/Yunji Noh/OneDrive/image_processing/data/BC2.jpg"
img3= "C:/Users/Yunji Noh/OneDrive/image_processing/data/BC3.jpg"
img4 = "C:/Users/Yunji Noh/OneDrive/image_processing/data/BC4.jpg"
img5= "C:/Users/Yunji Noh/OneDrive/image_processing/data/BC5.jpg"
img6 = "C:/Users/Yunji Noh/OneDrive/image_processing/data/BC6.jpg"
img7= "C:/Users/Yunji Noh/OneDrive/image_processing/data/BC7.jpg"
img8= "C:/Users/Yunji Noh/OneDrive/image_processing/data/BC8.jpg"
img9 = "C:/Users/Yunji Noh/OneDrive/image_processing/data/BC9.jpg"
img10 ="C:/Users/Yunji Noh/OneDrive/image_processing/data/BC10.jpg"
img11 = "C:/Users/Yunji Noh/OneDrive/image_processing/data/BC11.jpg"                                                     
img12 = "C:/Users/Yunji Noh/OneDrive/image_processing/data/BC12.jpg"
img13 = "C:/Users/Yunji Noh/OneDrive/image_processing/data/BC13.jpg"
img14 = "C:/Users/Yunji Noh/OneDrive/image_processing/data/BC14.jpg"
img15 = "C:/Users/Yunji Noh/OneDrive/image_processing/data/BC15.jpg"
img16 = "C:/Users/Yunji Noh/OneDrive/image_processing/data/BC16.jpg"
img17 = "C:/Users/Yunji Noh/OneDrive/image_processing/data/BC17.jpg"
img18 = "C:/Users/Yunji Noh/OneDrive/image_processing/data/BC18.jpg"
img19 = "C:/Users/Yunji Noh/OneDrive/image_processing/data/BC19.jpg"
img20 = "C:/Users/Yunji Noh/OneDrive/image_processing/data/BC20.jpg"
img21 = "C:/Users/Yunji Noh/OneDrive/image_processing/data/BC21.jpg"

pictures = [img1, img2, img3, img4, img5, img6, img7, img8, img9, img10, img11, img12, img13, img14, img15, img16, img17, img18, img19, img20, img21]

#사진이 0번 인덱스부터 시작이라 "원하는 사진 - 1"
img = cv2.imread(pictures[2])

# 그레이스케일 변환
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# 블러 적용
blur = cv2.medianBlur(gray, 9)

# 캐니 엣지 감지 수행
edges = cv2.Canny(blur, 100 , 120)

# 닫기 연산을 위한 구조 요소 생성
kernel = np.ones((11, 11), np.uint8)

# 닫기 연산 수행
closed = cv2.morphologyEx(edges, cv2.MORPH_CLOSE, kernel)

cv2.imshow('blur' , blur)
cv2.imshow('canny', edges)
cv2.imshow('mopology', closed)
cv2.waitKey(0)
cv2.destroyAllWindows()
############################################################

# 윤곽선 찾기
contours, _ = cv2.findContours(closed, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

# 새로운 캔버스 생성
canvas = np.zeros_like(img)

# 면적이 원본 이미지의 10% 이상인 윤곽선 그리기
for contour in contours:
    area = cv2.contourArea(contour)
    if area > 0.1 * img.shape[0] * img.shape[1]:
        # 윤곽선 근사화
        epsilon = 0.06 * cv2.arcLength(contour, True)
        approx = cv2.approxPolyDP(contour, epsilon, True)
        cv2.drawContours(canvas, [approx], -1, (0, 255, 0), 2)

# 결과 출력
cv2.imshow('Contours', canvas)
cv2.waitKey(0)
cv2.destroyAllWindows()

if len(approx) == 4:
    # 사각형의 네 꼭지점 좌표 추출
    points = approx.reshape(-1, 2).astype(np.float32)

    #명함 출력 크기 (단, 세로명함은 w = 300, h= 400 / 가로명함 w =400, h=200)
    w = 400
    h = 200

    # 변환 후 좌표 설정 (수동으로 방향 돌려주기....)
    final_points = np.array([[0, 0],[0, h],[w, h],[w, 0]], dtype=np.float32)

    # 변환 행렬 계산
    transform_matrix = cv2.getPerspectiveTransform(points, final_points)

    # 이미지 변환
    final_image = cv2.warpPerspective(img, transform_matrix, (w, h), flags=cv2.INTER_CUBIC)
    #sharp = cv2.filter2D(final_image, -1, np.array([[-1, -1, -1], [-1, 9, -1], [-1, -1, -1]]))

    # 결과 출력
    cv2.imshow('sharp_final', final_image)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
```

```python
import cv2
import numpy as np


img1 = "C:/Users/Yunji Noh/OneDrive/image_processing/project_img/BC1.jpg"
img2= "C:/Users/Yunji Noh/OneDrive/image_processing/project_img/BC2.jpg"
img3= "C:/Users/Yunji Noh/OneDrive/image_processing/project_img/BC3.jpg"
img4 = "C:/Users/Yunji Noh/OneDrive/image_processing/project_img/BC4.jpg"
img5= "C:/Users/Yunji Noh/OneDrive/image_processing/project_img/BC5.jpg"
img6 = "C:/Users/Yunji Noh/OneDrive/image_processing/project_img/BC6.jpg"
img7= "C:/Users/Yunji Noh/OneDrive/image_processing/project_img/BC7.jpg"
img8= "C:/Users/Yunji Noh/OneDrive/image_processing/project_img/BC8.jpg"
img9 = "C:/Users/Yunji Noh/OneDrive/image_processing/project_img/BC9.jpg"
img10 ="C:/Users/Yunji Noh/OneDrive/image_processing/project_img/BC10.jpg"
img11 = "C:/Users/Yunji Noh/OneDrive/image_processing/project_img/BC11.jpg"                                                     
img12 = "C:/Users/Yunji Noh/OneDrive/image_processing/project_img/BC12.jpg"
img13 = "C:/Users/Yunji Noh/OneDrive/image_processing/project_img/BC13.jpg"
img14 = "C:/Users/Yunji Noh/OneDrive/image_processing/project_img/BC14.jpg"
img15 = "C:/Users/Yunji Noh/OneDrive/image_processing/project_img/BC15.jpg"
img16 = "C:/Users/Yunji Noh/OneDrive/image_processing/project_img/BC16.jpg"
img17 = "C:/Users/Yunji Noh/OneDrive/image_processing/project_img/BC17.jpg"
img18 = "C:/Users/Yunji Noh/OneDrive/image_processing/project_img/BC18.jpg"
img19 = "C:/Users/Yunji Noh/OneDrive/image_processing/project_img/BC19.jpg"
img20 = "C:/Users/Yunji Noh/OneDrive/image_processing/project_img/BC20.jpg"
img21 = "C:/Users/Yunji Noh/OneDrive/image_processing/project_img/BC21.jpg"

pictures = [img1, img2, img3, img4, img5, img6, img7, img8, img9, img10, img11, img12, img13, img14, img15, img16, img17, img18, img19, img20, img21]

#사진이 0번 인덱스부터 시작이라 "원하는 사진 - 1"
img = cv2.imread(pictures[18])

###################part2########################################
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
blur = cv2.medianBlur(img_gray, 9)


cv2.imshow('blur', blur)
cv2.waitKey(0)
cv2.destroyAllWindows()


###############################part3#####################################
#blur처리 후 엣지 강조를 위한 필터 ; sigma_s(필터 강도), sigma_r(엣지 강도)
blur_pre_edge = cv2.edgePreservingFilter(blur, flags = 1, sigma_s = 100, sigma_r = 0.0001)

###대비 조절
alpha = 1.5  # 대비를 증가시키기 위한 상수
beta = 0    # 밝기 조정을 위한 상수
adjusted_img = cv2.convertScaleAbs(blur_pre_edge, alpha=alpha, beta=beta)

# Sobel 필터 적용
sobel_x = cv2.Sobel(blur_pre_edge, cv2.CV_64F, 1, 0, ksize=3)
sobel_y = cv2.Sobel(blur_pre_edge, cv2.CV_64F, 0, 1, ksize=3)

# Sobel 결과를 양수로 변환
sobel_x = cv2.convertScaleAbs(sobel_x)
sobel_y = cv2.convertScaleAbs(sobel_y)

# Sobel X와 Sobel Y 결과를 결합하여 엣지 강도 계산
sobel = cv2.addWeighted(sobel_x, 0.5, sobel_y, 0.5, 0)

# 이진화 수행
_, binary = cv2.threshold(sobel, 0, 255, cv2.THRESH_BINARY+cv2.THRESH_OTSU)

# 닫기 연산을 위한 구조 요소 생성
kernel = np.ones((11, 11), np.uint8)

# 닫기 연산 수행
closed = cv2.morphologyEx(binary, cv2.MORPH_CLOSE, kernel)

cv2.imshow('blur_pre_edge' , blur_pre_edge)
cv2.imshow('sobel', sobel)
cv2.imshow('mopology', closed)
cv2.waitKey(0)
cv2.destroyAllWindows()

##########part4##########################################################

# 윤곽선 찾기
contours, _ = cv2.findContours(closed, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)


# 새로운 캔버스 생성
canvas = np.zeros_like(img)

# 면적이 원본 이미지의 10% 이상인 윤곽선 그리기
for contour in contours:
    area = cv2.contourArea(contour)
    if area > 0.1 * img.shape[0] * img.shape[1]:
        # 윤곽선 근사화
        epsilon = 0.06* cv2.arcLength(contour, True)
        approx = cv2.approxPolyDP(contour, epsilon, True)
        cv2.drawContours(canvas, [approx], -1, (0, 255, 0), 2)

# 결과 출력
cv2.imshow('Contours', canvas)
cv2.waitKey(0)
cv2.destroyAllWindows()


if len(approx) == 4:
    # 사각형의 네 꼭지점 좌표 추출
    points = approx.reshape(-1, 2).astype(np.float32)

    w = 400
    h = 300

    # 변환 후 좌표 설정 (수동으로 방향 돌려주기....)
    dst_points = np.array([[0, h],[w, h], [w, 0], [0, 0]], dtype=np.float32)

    # 변환 행렬 계산
    transform_matrix = cv2.getPerspectiveTransform(points, dst_points)

    # 이미지 변환
    dst_image = cv2.warpPerspective(img, transform_matrix, (w, h))
    sharp = cv2.filter2D(dst_image, -1, np.array([[-1, -1, -1], [-1, 9, -1], [-1, -1, -1]]))

    # 결과 출력
    cv2.imshow('sharp', sharp)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
```
