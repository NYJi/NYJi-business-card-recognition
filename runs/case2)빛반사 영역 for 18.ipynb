```python
import cv2
import numpy as np

# 이미지를 로드합니다.
img = cv2.imread("C:/Users/Yunji Noh/OneDrive/image_processing/data/BC18.jpg")
```

```python
def img_process(img):
    # 이미지를 그레이스케일로 변환
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    blur = cv2.GaussianBlur(gray, (11,11), 0)
    
    ret, dst = cv2.threshold(blur, 225, 240, cv2.THRESH_BINARY)
    
     # 윤곽선 찾기
    contours, _ = cv2.findContours(dst, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

    # 새로운 캔버스 생성
    canvas = np.zeros_like(img)

    # 면적이 원본 이미지의 10% 이상인 윤곽선 그리기
    for contour in contours:
        area = cv2.contourArea(contour)
        if area > 0.1* img.shape[0] * img.shape[1]:
            # 윤곽선 근사화
            epsilon = 0.01 * cv2.arcLength(contour, True)
            approx = cv2.approxPolyDP(contour, epsilon, True)
            cv2.drawContours(canvas, [approx], -1, (0, 255, 0), 2)
            
      # 원근 변환 적용
    src_pts = np.float32(approx)
    dst_pts = np.float32([[img.shape[1], img.shape[0]],[img.shape[1], 0],[0, 0], [0, img.shape[0]]])
    matrix = cv2.getPerspectiveTransform(src_pts, dst_pts)
    warped_image = cv2.warpPerspective(img, matrix, (img.shape[1], img.shape[0]))
    sharp = cv2.filter2D(warped_image, -1, np.array([[-1, -1, -1], [-1, 9, -1], [-1, -1, -1]]))
    
    
    # 이미지 크기 조정
    resized_image = cv2.resize(sharp, (400, 200))
    resized_img = cv2.resize(warped_image, (400, 200))
 
    
    #결과 폴더 저장
    #cv2.imwrite("C:/Users/Yunji Noh/OneDrive/image_processing/results/final_image(sharp(x))/BC18.jpg", resized_img)
    cv2.imwrite("C:/Users/Yunji Noh/OneDrive/image_processing/results/final_image/BC18.jpg", resized_img)
    # 결과 이미지 출력
    cv2.imshow('contours', canvas)
    cv2.imshow("final Image",  resized_image)
    cv2.imshow('binary', dst)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
```

```python
img_process(img)
```

```python
import cv2
import numpy as np
import random

img1 = "C:/Users/Yunji Noh/OneDrive/image_processing/project_img/BC1.jpg"
img2= "C:/Users/Yunji Noh/OneDrive/image_processing/project_img/BC2.jpg"
img3= "C:/Users/Yunji Noh/OneDrive/image_processing/project_img/BC3.jpg"
img4 = "C:/Users/Yunji Noh/OneDrive/image_processing/project_img/BC4.jpg"
img5= "C:/Users/Yunji Noh/OneDrive/image_processing/project_img/BC5.jpg"
img6 = "C:/Users/Yunji Noh/OneDrive/image_processing/project_img/BC6.jpg"
img7= "C:/Users/Yunji Noh/OneDrive/image_processing/project_img/BC7.jpg"
img8= "C:/Users/Yunji Noh/OneDrive/image_processing/project_img/BC8.jpg"
img9 = "C:/Users/Yunji Noh/OneDrive/image_processing/project_img/BC9.jpg"
img10 ="C:/Users/Yunji Noh/OneDrive/image_processing/project_img/BC10.jpg"
img11 = "C:/Users/Yunji Noh/OneDrive/image_processing/project_img/BC11.jpg"                                                     
img12 = "C:/Users/Yunji Noh/OneDrive/image_processing/project_img/BC12.jpg"
img13 = "C:/Users/Yunji Noh/OneDrive/image_processing/project_img/BC13.jpg"
img14 = "C:/Users/Yunji Noh/OneDrive/image_processing/project_img/BC14.jpg"
img15 = "C:/Users/Yunji Noh/OneDrive/image_processing/project_img/BC15.jpg"
img16 = "C:/Users/Yunji Noh/OneDrive/image_processing/project_img/BC16.jpg"
img17 = "C:/Users/Yunji Noh/OneDrive/image_processing/project_img/BC17.jpg"
img18 = "C:/Users/Yunji Noh/OneDrive/image_processing/project_img/BC18.jpg"
img19 = "C:/Users/Yunji Noh/OneDrive/image_processing/project_img/BC19.jpg"
img20 = "C:/Users/Yunji Noh/OneDrive/image_processing/project_img/BC20.jpg"
img21 = "C:/Users/Yunji Noh/OneDrive/image_processing/project_img/BC21.jpg"


pictures = [img1, img2, img3, img4, img5, img6, img7, img8, img9, img10, img11, img12, img13, img14, img15, img16, img17, img18, img19, img20, img21]

# 원하는 사진 선택 (0번 인덱스부터 시작하므로 원하는 사진 - 1)
img= cv2.imread(pictures[17])

# 이미지를 그레이스케일로 변환
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# 밝기 임계값 설정
threshold1 = 230  # 조명 영역을 판별하기 위한 임계값
#threshold2 = 
# 임계값 이상인 픽셀을 어두운 색(0, 0, 0)으로 변환
dark_pixels = gray <  threshold1
img[dark_pixels] = (100, 100, 100)

# blur 처리
blur = cv2.GaussianBlur(img, (17, 17), 0)

edge = cv2.Canny(blur, 120, 200)

_, src_bin = cv2.threshold(edge, 0, 255, cv2.THRESH_BINARY | cv2.THRESH_OTSU)

kernel = np.ones((5, 5), np.uint8)
closed = cv2.morphologyEx(src_bin, cv2.MORPH_CLOSE, kernel)


# 외곽선 검출
contours, _ = cv2.findContours(closed, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

# 가장 외곽에 있는 외곽선 선택
outer_contour = max(contours, key=cv2.contourArea)

# 외곽선을 그리기 위한 캔버스 생성
canvas = np.zeros_like(img)

# 외곽선 그리기
cv2.drawContours(canvas, [outer_contour], -1, (0, 0, 255), 2)


# 꼭짓점 추출
epsilon = 0.02 * cv2.arcLength(outer_contour, True)
approx = cv2.approxPolyDP(outer_contour, epsilon, True)
approx = approx.reshape(-1, 2)  # (4, 1, 2) 형태를 (4, 2)로 변환

w = 400
h = 300

# 변환된 이미지에서의 꼭짓점 좌표
dst_points = np.float32([[0, 0], [0, h], [w, h],[w, 0] ])

# 원근 변환 행렬 계산
perspective_matrix, _ = cv2.findHomography(approx, dst_points)

# 원근 변환 적용
warped = cv2.warpPerspective(img, perspective_matrix, (w, h))

# 결과 출력
cv2.imshow('Outer Contour', canvas)
cv2.imshow('Warped Image', warped)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
